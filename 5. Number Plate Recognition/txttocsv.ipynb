```python
import pandas as pd
import os 
from glob import glob
```

```python
os.getcwd()
```

```python
path = glob('./Labels/*.txt')
```

```python
filename = path[0]
cordinates = open(filename,'r').read().split(' ')
cordinates
```

```python
train_data = dict(ImageName=[], FileName=[],Class=[],
x_center=[], y_center=[],
width=[],height=[])

for filename in path:
    cordinates = open(filename,'r').read().split(' ')
    
    train_data['FileName'].append(filename)
    train_data['Class'].append(cordinates[0])
    train_data['x_center'].append(cordinates[1])
    train_data['y_center'].append(cordinates[2])
    train_data['width'].append(cordinates[3])
    train_data['height'].append(cordinates[4])
```

```python
train_data = pd.DataFrame(train_data)
```

```python
train_data.to_csv('train_data.csv',index=False)
```
